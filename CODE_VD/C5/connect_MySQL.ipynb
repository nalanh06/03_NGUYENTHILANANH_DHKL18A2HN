```python
import mysql.connector
from mysql.connector import Error

# THAM SỐ KẾT NỐI (Phải khớp với thiết lập Docker)
HOST = "127.0.0.1" # Host: Máy tính cục bộ của bạn
USER = "root"      # User: Tài khoản root mặc định của Docker MySQL
PASSWORD = "uneti" # Password: Mật khẩu bạn đã thiết lập bằng tham số -e
DATABASE = "data_science_db" # DB: Database bạn đã tạo sẵn
PORT = 3307        # PORT: Cổng bạn đã sửa để tránh xung đột

conn = None # Khởi tạo kết nối

try:
    print("Đang cố gắng kết nối tới MySQL Server...")
    # 1. Thực hiện kết nối
    conn = mysql.connector.connect(
        host=HOST,
        user=USER,
        password=PASSWORD,
        database=DATABASE,
        port=PORT
    )

    # 2. Kiểm tra trạng thái kết nối
    if conn.is_connected():
        print(f"\n✅ Kết nối THÀNH CÔNG tới CSDL '{DATABASE}' qua cổng {PORT}.")

        # 3. Tạo cursor và thực hiện truy vấn mẫu đơn giản
        cursor = conn.cursor()
        print("\nThực hiện truy vấn mẫu...")
        cursor.execute("SELECT VERSION(), CURRENT_DATE()")
        result = cursor.fetchone()

        print(f"Phiên bản MySQL: {result[0]}")
        print(f"Ngày hiện tại trên Server: {result[1]}")

    else:
        print("❌ Kết nối THẤT BẠI.")

except Error as e:
    print(f"\n❌ Lỗi kết nối MySQL: {e}")

finally:
    # 4. Đảm bảo đóng kết nối an toàn
    if conn and conn.is_connected():
        cursor.close()
        conn.close()
        print("\nĐã đóng kết nối CSDL.")
```

```python
import sqlite3 
 
# Kết nối đến cơ sở dữ liệu (hoặc tạo cơ sở dữ liệu mới nếu nó không tồn tại) 
conn = sqlite3.connect('mydatabase.db') 
 
# Tạo một đối tượng con trỏ (Cursor object) từ đối tượng kết nối 
cursor = conn.cursor() 
 
# Thực thi câu lệnh SQL và truy xuất dữ liệu được đặt trong cặp dấu '''...''' 
cursor.execute(''' 
               CREATE TABLE IF NOT EXISTS users ( 
               id INTEGER PRIMARY KEY,  
               "Tên" TEXT,  
               "Tuổi" INTEGER 
               ) 
               ''') 
cursor.execute(''' 
               INSERT INTO users ("Tên", "Tuổi")  
               VALUES ("Nguyễn Văn A", 30) 
               ''') 
 
# Lưu các thay đổi vào cơ sở dữ liệu (nếu cần) 
conn.commit() 
 
# Truy xuất dữ liệu từ bảng users 
cursor.execute("SELECT * FROM users") 
rows = cursor.fetchall() 
 
# In kết quả 
for row in rows: 
    print(row) 
 
# Đóng con trỏ (Cursor object) và đóng kết nối 
cursor.close() 
conn.close() 
```

```python
# code for update operation  
import sqlite3  
   
# Kết nối đến cơ sở dữ liệu, tên cơ sở dữ lệu được truyền như một tham số của connect() 
conn = sqlite3.connect('mydatabase.db')  
   
# cập nhật bản ghi nhân viên từ bảng users  
conn.execute("UPDATE users SET 'Tên' = 'Nguyễn Vân Anh' where id='1'")  
conn.commit()  
   
print("Tổng số dòng được cập nhật :", conn.total_changes)  
   
cursor = conn.execute("SELECT * FROM users")  
for row in cursor:  
   print(row)  
# Đóng kết nối sau khi hoàn tất công việc  
conn.close()
```
