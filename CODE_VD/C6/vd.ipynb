```python
import threading
def print_cube(num):
# function to print cube of given num
    print("Cube: {}".format(num * num * num))
def print_square(num):
# function to print square of given num
    print("Square: {}".format(num * num))
if __name__ == "__main__":
# creating thread
    t1 = threading.Thread(target=print_square, args=(10,))
    t2 = threading.Thread(target=print_cube, args=(10,))
    # start the threads
    t1.start()
    t2.start()
    # wait until both threads finish
    t1.join()
    t2.join()
    # continue with the main loop
    for i in range(5):
        print("Main loop: {}".format(i))
    print("Done!")
```

```python
import threading
class PrintThread(threading.Thread):
    def __init__(self, num):
        super().__init__()
        self.num = num
    def run(self):
        self.print_square()
        self.print_cube()
    def print_cube(self):
        # function to print cube of given num
        print("Cube: {}".format(self.num * self.num * self.num))
    def print_square(self):
        # function to print square of given num
        print("Square: {}".format(self.num * self.num))
if __name__ == "__main__":
    # creating thread
    t = PrintThread(10)
    # start the thread
    t.start()
    # wait until the thread finishes
    t.join()
    # continue with the main loop
    for i in range(5):
        print("Main loop: {}".format(i))
    print("Done!")
```

```python
from concurrent.futures import ThreadPoolExecutor
import pandas as pd
import time
def load_csv(filename):
    print(f"Đang đọc {filename}")
    time.sleep(0.5)
    df=pd.read_csv(filename)
    return filename, len(df)
files = ["sale_jan.csv","sales_feb.csv","sales_mar.csv"]
with ThreadPoolExecutor(max_workers=3) as executor:
    results=executor.map(load_csv,files)
for fname, rows in results:
    print(f"{fname}: {rows} dòng")
```

```python
import time
import pandas as pd

files = ["sales_jan.csv","sales_feb.csv","sales_mar.csv"]

def read_file_sequential():
    start = time.time()
    
    for f in files:
        time.sleep(0.2)
        df = pd.read_csv(f)
        print(f"Đọc xong {f}: {len(df)} dòng")
    end = time.time()
    print(f"Thời gian đọc tuần tự: {end-start:.4f} giây")
```

```python
#B đọc file bằng threading.Thread
import threading
import time
import pandas as pd
files = ["sales_jan.csv","sales_feb.csv","sales_mar.csv"]
results=[]
def load_csv_thread(filename):
    time.sleep(0.2)
    df=pd.read_csv(filename)
    results.append((filename, len(df)))
    print(f"Đọc xong {filename}: {len(df)} dòng")
def read_file_multithread():
    start = time.time()
    threads=[]
    for f in files:
        t=threading.Thread(target=load_csv_thread, args=(f,))
        threads.append(t)
        t.start()
    for t in threads:
        t.join()
    end=time.time()
    print(f"Thời gian đọc đa luồng: {end-start:.4f} giây") 
```

```python
import threading
# global variable x
x = 0
#Viết hàm increment(), tăng giá trị biến toàn cục x
def increment():
    global x
    tmp=x
    time.sleep(0.000001)
    x =tmp+ 1
#Hàm theread_task() thực hiện tác vụ luồng, thực hiện gọi hàm
# increment() 100000 lần.'''
def thread_task():
    for _ in range(100000):
        increment()
#Hàm main_task(): thực hiện nhiệm vụ chính của chương trình.
#Nó đặt giá trị của biến toàn cục x=0, tạo hai luồng (t1 và t2)
# để thực thi công việc trong thread_task().
def main_task():
    global x
    x = 0
    # creating threads
    t1 = threading.Thread(target=thread_task)
    t2 = threading.Thread(target=thread_task)
    # start threads
    t1.start()
    t2.start()
    # wait until threads finish their job
    t1.join()
    t2.join()
#Chương trình chạy vòng lặp 10 lần,
#trong mỗi lần lặp gọi main_task() để thực hiện công việc.
#Sau đó in giá trị x ở mỗi lần lặp ra màn hình.
for i in range(10):
    main_task()
    print("Lần lặp {0}: x = {1}".format(i, x))
```

```python
import threading
import time
balance=100
def add_money():
    global balance
    tmp=balance
    tmp=balance
    time.sleep(0.001)
    balance=tmp+10
threads=[threading.Thread(target=add_money) for _ in range(2)]
for t in threads:
    t.start()
for t in threads:
    t.join()
print("Số dư cuối: ", balance)
```

```python
import threading
import time
lock=threading.Lock()
balcance=100
def add_money_safe():
    global balance
    lock.acquire()
    try:
        tmp=balance
        time.sleep(0.001)
        balance=tmp+10
    finally:
        lock.release()
threads=[threading.Thread(target=add_money) for _ in range(2)]
for t in threads:
    t.start()
for t in threads:
    t.join()
print("Số dư cuối: ", balance)
```

```python
from threading import Lock
import time
lock=Lock()
balance=100
def add_money():
    global balance
    with lock:
        tmp=balance
        time.sleep(0.001)
        balance=tmp+10
threads=[threading.Thread(target=add_money) for _ in range(2)]
for t in threads:
    t.start()
for t in threads:
    t.join()
print("Số dư cuối: ", balance)
```

```python
import threading
import time
sem=threading.Semaphore(3)
def access(id):
    sem.acquire()
    print(f"Thread {id} bắt đầu truy cập")
    time.sleep(1)
    print(f"Thread {id} kết thúc")
    sem.release()
for i in range(6):
    threading.Thread(target=access, args=(i,)).start()
```

```python
import threading
import time
event= threading.Event()
def worker():
    print("Worker chờ tín hiệu...")
    event.wait()
    print("Worker bắt đầu làm việc")
def controller():
    time.sleep(1)
    print("Contruller phát tín hiệu")
    event.set()
threading.Thread(target=worker).start()
threading.Thread(target=controller).start()
```

```python
import queue
import threading
import time
q=queue.PriorityQueue()
def producer():
    for i in range(5):
        q.put((i, f"task {i}"))
        print("Producer đưa vào: ",i)
        time.sleep(0.5)
def consumer():
    while True:
        item=q.get()
        print("Consumer xử lý:",item)
        q.task_done()
threading.Thread(target=consumer, daemon=True).start()
threading.Thread(target=producer).start()    
```
