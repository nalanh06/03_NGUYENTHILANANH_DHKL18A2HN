```python
import pandas as pd
import numpy as np

ser = pd.Series(np.random.normal(10, 5, 25))
```

Giá trị tối thiểu (mininum) của ser

```python
min_value = ser.min()

print("Giá trị tối thiểu của ser:", min_value)
```

 Phần centile thứ 25 (25th percentile) của ser. Đây là giá trị mà 25% các giá trị trong chuỗi nhỏ hoặc bằng nó

```python
sorted_ser = ser.sort_values()
index_25th_percentile = int(len(sorted_ser) * 0.25)
percentile_25 = sorted_ser.iloc[index_25th_percentile]

print("Phần centile thứ 25 của ser:", percentile_25)
```

Trung vị (median) của ser. Đây là giá trị ở vị trí giữa khi sắp xếp ser theo thứ tự tăng dần

```python
median_value = ser.median()

print("Trung vị của ser:", median_value)
```

Giá trị tối đa (maxinum) của ser

```python
max_value = ser.max()

print("Giá trị tối đa của ser:", max_value)
```
